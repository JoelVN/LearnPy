
In addition to the above comparison operator Python uses:

is: Returns true if both variables are the same object(x is y)
is not: Returns true if both variables are not the same object(x is not y)
in: Returns True if the queried list contains a certain item(x in y)
not in: Returns True if the queried list doesn't have a certain item(x in y)


```python

print('1 is 1', 1 is 1)                   # True - because the data values are the same
print('1 is not 2', 1 is not 2)           # True - because 1 is not 2
print('A in Asabeneh', 'A' in 'Asabeneh') # True - A found in the string
print('B in Asabeneh', 'B' in 'Asabeneh') # False - there is no uppercase B
print('coding' in 'coding for all') 
```

Slope is (m = y2-y1/x2-x1). Find the slope and Euclidean distance between point (2, 2) and point (6,10)

```python
# puntos (2, 2) and  (6,10)
x1,y1 = 2,2
x2,y2 = 6,10

#calcular pendiente (slope)
slope = (y2 - y1) / (x2 - x1)

# distancia euclidean
distancia = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

# Imprimir los resultados
print("Pendiente:", slope)
print("Distancia Euclidiana:", distancia)
```

Calculate the value of y (y = x^2 + 6x + 9). Try to use different x values and figure out at what x value y is going to be 0.

```python
# Definir la variable simbólica x y la ecuación
x = sp.symbols('x')
ecuacion = x**2 + 6*x + 9

# Resolver la ecuación para encontrar las raíces (valores de x donde y = 0)
raices = sp.solve(ecuacion, x)

# Imprimir las raíces
print("Las raíces son:", raices)

# Calcular los valores de y para las raíces encontradas
valores_de_y = [ecuacion.subs(x, raiz) for raiz in raices]

# Imprimir los valores de y correspondientes a las raíces
for i, raiz in enumerate(raices):
    print(f"Para x = {raiz}, y = {valores_de_y[i]}")
```
